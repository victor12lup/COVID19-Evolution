## Data 

Data has been retreived from "https://www.kaggle.com/imdevskp/corona-virus-report" where data from confirmed cases and deaths from most of the countries in the world has been collected from the Johns Hopkins University database. This data has been constanlty updated (last day included is 27 of April of 2021). 4 data sets can be dowloaded however in this notebook only 2 are considered. One has the cumulative number of confirmed cases and the other has confirmed deaths.   

```python
df = pd.read_csv("time_series_covid_19_confirmed.csv",  parse_dates=True)
df1 =  pd.read_csv("time_series_covid_19_deaths.csv")
```

From the data set we will only use the cases reported , therefore all the other variables included in the datasets are dropped

```python
df = df.drop(['Province/State','Lat','Long'], axis=1)
df1 =  df1.drop(['Province/State','Lat','Long'], axis=1)

#Countries seem to have multiple entries representing different areas within a same country therefore they are grouped
df = df.groupby('Country/Region').sum()
df1 = df1.groupby("Country/Region").sum()
```

```python
#columns are transfromed into datetime format to help with the analysis and transformation
df.columns = df.columns.map(date)
df1.columns = df1.columns.map(date)
```

It can now be appreciated the general format of the dataframes with the total information

```python
display(df)
display(df1)
```

```python
# making a copy of the dataframe to have both cumulative and daily increases
df_cumu =df.copy()
df1_cumu =df1.copy()
#Use the function daily_increase to change from cumulative to daily increase
df = daily_increase(df,df_cumu)
df1 = daily_increase(df1,df1_cumu)
```

```python
#some countries have reported lower cases from one day to another this is corrected by setting a negative value to Nan and then
#filling it with the cases from a day before
df[df<0] = None
df= df.fillna(method='bfill',axis=1)
df1[df1<0] = None
df1= df1.fillna(method='bfill',axis=1)
# After analysing the graphs Turkey has an outlier and India as well , this may be an error in the reporting of the number so it is
# corrected with the actual number found on the internet
df['2020-12-10'].loc['Turkey'] = 30424
df['2021-04-27'].loc['India'] = 362902
df1['2021-04-27'].loc['India']= 3285
```

From all the countries is useful to make a dictionary in order to separate them by continent or groups as top 10 or biggest population (values taken from the internet)

```python
from collections import defaultdict
#Creating a default dictionary in order to have different continets with lists of countries
continents = defaultdict(list)
# Some countries have different names that are not recognized by the module pycountry_convert therefore they needd to be added manually
diff= []

countries= list(monthly(df_cumu).columns)
for x in countries:
    continents["All"].append(x)
    try:
        if country_to_continent(x) == 'Asia':
            continents[country_to_continent(x)].append(x)
        if country_to_continent(x) == 'Europe':
            continents[country_to_continent(x)].append(x)
        if country_to_continent(x) == 'North America':
            continents[country_to_continent(x)].append(x)
        if country_to_continent(x) == 'South America':
            continents[country_to_continent(x)].append(x)
        if country_to_continent(x) == 'Oceania':
            continents[country_to_continent(x)].append(x)
        if country_to_continent(x) == 'Africa':
            continents[country_to_continent(x)].append(x)
    
    # skip the error we already know is beacuse the names have different wording
    except KeyError :
        diff.append(x)
    except TypeError :
        diff.append(x)
    
# manually adding the different country names 
continents['Asia'].append(diff[0])
continents['Asia'].append(diff[9])
continents['Africa'].append(diff[1])
continents['Africa'].append(diff[2])
continents['Africa'].append(diff[3])
continents['North America'].append(diff[11])
continents['Most populated'] = ['China','India','US','Indonesia','Pakistan','Brazil','Nigeria','Bangladesh','Russia','Mexico']
continents['Top 10'] = ['US','India','Brazil','Russia','United Kingdom','France','Spain','Italy','Turkey','Germany']
```
